```python
import pandas as pd

df = pd.read_csv("../DATA/combined_data.csv")
```

```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

df['year'] = pd.to_datetime(df['year'], format='%Y')
df.set_index('year', inplace=True)
gdp = df['GDP']
model = ARIMA(gdp, order=(1, 1, 1))  # Adjust the order as needed
model_fit = model.fit()
```

```python
print(model_fit.summary())
```

```python
unemployment = df['Unemployment_rate']
model_unemployment = ARIMA(unemployment, order=(1, 1, 1))  # Adjust the order
model_fit_unemployment = model_unemployment.fit()
print(model_fit_unemployment.summary())
```

```python
immigration = df['num_immigrants']
model_immigration = ARIMA(immigration, order=(1, 1, 1))  # Adjust the order
model_fit_immigration = model_immigration.fit()
print(model_fit_immigration.summary())
```

```python
plt.figure(figsize=(12, 6))
plt.plot(gdp, label='Original GDP')
plt.plot(model_fit.fittedvalues, label='Fitted Values')
plt.title('ARIMA Model Fit')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(unemployment, label='Original Unemployment')
plt.plot(model_fit_unemployment.fittedvalues, label='Fitted Values')
plt.title('ARIMA Model Fit')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(immigration, label='Original Immigration')
plt.plot(model_fit_immigration.fittedvalues, label='Fitted Values')
plt.title('ARIMA Model Fit')
plt.legend()
plt.show()
```
